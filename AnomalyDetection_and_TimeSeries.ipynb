#### #Question 1 : What is Dimensionality Reduction? Why is it important in machine learning?
##### #Ans.Dimensionality Reduction is a process used in machine learning and data analysis to reduce the number of input variables (features) in a dataset while retaining as much important information as possible.In simple terms, it means converting high-dimensional data (many features) into low-dimensional data (fewer features) without losing too much meaningful information.
Example:
Suppose you have a dataset of students with 100 features like height, weight, grades, attendance, hobbies, etc.
Many of these features may be correlated or redundant.
Dimensionality reduction helps you compress this into maybe 2 or 3 key features (or combinations of them) that still describe the data well.

Why It's Important:
1. Removes noise and redundancy:\
Eliminates unnecessary features that don't add value.
2. Reduces computation cost:\
Fewer features mean faster training and prediction.
3. Avoids overfitting:\
Simplifies the model, reducing the risk of memorizing noise.
4. Improves visualization:\
Enables plotting and understanding of high-dimensional data in 2D or 3D.
5. Enhances model performance:\
Helps models generalize better by focusing on key patterns.


#### #Question 2: Name and briefly describe three common dimensionality reduction techniques..
##### #Ans.
1. Principal Component Analysis (PCA):
PCA transforms the original features into new uncorrelated variables called principal components, which capture the maximum variance in the data. It's the most commonly used linear dimensionality reduction technique.

2. Linear Discriminant Analysis (LDA):
LDA reduces dimensions by finding feature combinations that best separate different classes in the data, making it useful for classification problems.

3. t-Distributed Stochastic Neighbor Embedding (t-SNE):
t-SNE is a non-linear technique mainly used for visualizing high-dimensional data in 2D or 3D while preserving the relationships between data points.

#### #Question 3: What is clustering in unsupervised learning? Mention three popular clustering algorithms.
##### #Ans.Clustering in unsupervised learning is the process of grouping similar data points together based on their features, without using any predefined labels. The goal is to ensure that data points within the same cluster are more similar to each other than to those in other clusters.

Three popular clustering algorithms are:
1. K-Means Clustering: Divides data into K clusters by minimizing the distance between points and their cluster centroids.
2. Hierarchical Clustering: Builds a tree-like structure of clusters by merging or splitting them step by step.
3. DBSCAN (Density-Based Spatial Clustering of Applications with Noise): Groups data points that are close in density and identifies outliers as noise.

#### #Question 4: Explain the concept of anomaly detection and its significance.
##### #Ans.Anomaly detection is the process of identifying unusual patterns or data points that deviate significantly from the normal behavior of a dataset. These unusual observations are called anomalies or outliers.

Significance:

1. Fraud Detection: Helps identify fraudulent transactions in banking or credit card data.
2. Network Security: Detects suspicious activities or potential cyberattacks.
3. Fault Detection: Finds errors or malfunctions in industrial systems and sensors.
4. Data Quality Improvement: Helps clean datasets by removing incorrect or noisy data.

#### #Question 5: List and briefly describe three types of anomaly detection techniques.
##### #Ans.
1. Statistical Methods:
These techniques assume that normal data follows a specific statistical distribution. Data points that deviate significantly from this distribution are flagged as anomalies.
Example: Z-score, Gaussian models.

2. Machine Learning-Based Methods:
These use algorithms to learn patterns of normal behavior and detect outliers that don't fit the learned model.
Example: Isolation Forest, One-Class SVM.

3. Distance or Density-Based Methods:
These methods detect anomalies by measuring the distance or density of data points. Points far from others or in low-density regions are considered anomalies.
Example: k-Nearest Neighbors (kNN), DBSCAN.


#### #Question 6: What is time series analysis? Mention two key components of time series data.
##### #Ans.Time Series Analysis is a statistical technique used to analyze data points collected or recorded over time at regular intervals, such as daily, monthly, or yearly. The goal is to understand the underlying patterns, trends, and relationships in the data, and to forecast future values based on past observations.Time series analysis is widely used in various fields such as finance (stock prices), economics (GDP, inflation), weather forecasting, sales prediction, and sensor monitoring. It helps in identifying consistent patterns and making data-driven decisions over time.

Two key components of time series data are:

1. Trend:
The long-term direction in which the data is moving over time — it can be upward, downward, or stable. For example, the steady increase in global temperature over years shows an upward trend.
2. Seasonality:
The repeating pattern or fluctuation that occurs at fixed intervals due to seasonal factors. For example, ice cream sales increase during summer and decrease during winter.

#### #Question 7: Describe the difference between seasonality and cyclic behavior in time series.
##### #Ans.Seasonality and cyclic behavior are two patterns commonly found in time series data, but they differ in their duration and causes.

1. Seasonality:
Refers to patterns that repeat at fixed and regular intervals due to seasonal factors such as weather, festivals, or business cycles.
* Example: Ice cream sales increasing every summer or electricity usage rising every winter.
It has a fixed and known period (e.g., every month, quarter, or year).

2. Cyclic Behavior:
Refers to long-term fluctuations that occur at irregular and variable intervals, often influenced by economic or business conditions.
* Example: Economic expansions and recessions that repeat over several years.
It does not have a fixed period — the duration and intensity of cycles vary.

```python
# Question 8: Write Python code to perform K-means clustering on a sample dataset.
# (Include your Python code and output in the code box below.)

import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=200, centers=3, cluster_std=0.70, random_state=42)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

labels = kmeans.labels_
centroids = kmeans.cluster_centers_

plt.figure(figsize=(7, 5))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title("K-Means Clustering Results", fontsize=13)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

#### #Question 9: What is inheritance in OOP? Provide a simple example in Python.
##### #Ans.Inheritance in Object-Oriented Programming (OOP) is a concept where a new class (called the child or derived class) acquires the properties and behaviors (attributes and methods) of an existing class (called the parent or base class).It promotes code reusability and helps in creating a hierarchy between classes.

Example:


```python
class Animal:
    def speak(self):
        print("Animals make sounds")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

dog1 = Dog()
dog1.speak()
```

#### #Question 10: How can time series analysis be used for anomaly detection?
##### #Ans.Time series analysis can be used for anomaly detection by identifying data points or patterns that deviate significantly from the expected temporal behavior. Since time series data is collected over time, anomalies often appear as sudden spikes, drops, or irregular patterns that break the normal trend or seasonality.

Ways it is used:
1. Statistical Methods:
Detect anomalies by comparing current values with predicted ranges using models like ARIMA or moving averages.
2. Machine Learning Models:
Use algorithms such as LSTM (Long Short-Term Memory) networks or Autoencoders to learn normal temporal patterns and flag unusual deviations.
3. Threshold-Based Detection:
Set upper and lower limits based on historical data — any value outside these bounds is considered an anomaly.

* Example:
Detecting sudden drops in website traffic, unexpected spikes in power consumption, or unusual sensor readings in IoT systems.
In short, time series analysis helps monitor and detect irregular behavior over time, improving reliability and early fault detection.
